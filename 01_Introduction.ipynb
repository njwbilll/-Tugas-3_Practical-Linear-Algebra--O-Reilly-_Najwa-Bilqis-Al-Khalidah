## 1.1 Apa itu Aljabar Linear dan Mengapa Mempelajarinya?

Aljabar linear adalah cabang matematika yang mempelajari matriks, vektor,
dan operasi-operasi yang bekerja pada keduanya.

### Sejarah Singkat

Aljabar linear memiliki sejarah panjang sejak abad ke-17 di Barat dan bahkan
lebih awal di Tiongkok. Matriks awalnya digunakan oleh Descartes untuk
menyimpan koordinat geometri, dan oleh Gauss untuk memecahkan sistem persamaan.

### "Soft Proof" melalui Kode

Pendekatan unik buku ini: asumsikan Python dan NumPy mengimplementasikan
komputasi dengan benar, lalu eksplorasi prinsip matematika melalui
banyak contoh numerik.

Contoh: membuktikan sifat komutatif perkalian (a * b = b * a)
melalui kode, bukan melalui bukti formal.

````python
import numpy as np

# "Soft proof" sifat komutatif perkalian skalar
# Kita verifikasi bahwa a * b = b * a untuk banyak angka acak

print("=== Soft Proof: Sifat Komutatif Perkalian ===")
print()

results_all_zero = True
for trial in range(10):
    a = np.random.randn()
    b = np.random.randn()
    diff = a * b - b * a
    print(f"Trial {trial+1:2d}: a={a:6.3f}  b={b:6.3f}  a*b - b*a = {diff:.2e}")
    if abs(diff) > 1e-10:
        results_all_zero = False

print()
if results_all_zero:
    print("Selisih selalu nol -- sifat komutatif terbukti secara numerik!")
else:
    print("Ada selisih non-nol -- ada masalah!")
````

````python
# Demonstrasi: pendekatan "soft proof" untuk sifat distributif
# Kita verifikasi bahwa a * (b + c) = a*b + a*c

print("=== Soft Proof: Sifat Distributif Perkalian ===")
print()

for trial in range(5):
    a = np.random.randn()
    b = np.random.randn()
    c = np.random.randn()

    lhs = a * (b + c)
    rhs = a * b + a * c
    diff = lhs - rhs

    print(f"Trial {trial+1}: a*(b+c) = {lhs:.6f}  |  a*b + a*c = {rhs:.6f}  |  selisih = {diff:.2e}")

print()
print("Selisih selalu mendekati nol (dalam presisi mesin floating point).")
print("Ini memverifikasi sifat distributif secara numerik.")
````

````python
# Demonstrasi pentingnya aljabar linear dalam data science
# Contoh sederhana: regresi linear menggunakan operasi matriks

print("=== Contoh: Aljabar Linear dalam Regresi Linear ===")
print()
print("Regresi linear adalah salah satu algoritma ML yang paling dasar.")
print("Di balik layar, ia menggunakan operasi matriks secara ekstensif.")
print()

# Buat data sederhana
np.random.seed(42)
n = 50
x = np.linspace(0, 10, n)
y = 2.5 * x + 1.0 + np.random.randn(n) * 1.5

# Buat matriks desain (design matrix)
A = np.column_stack([np.ones(n), x])

print(f"Data: {n} titik")
print(f"Matriks desain A shape: {A.shape}")
print(f"5 baris pertama matriks desain:")
print(A[:5])
print()

# Hitung koefisien menggunakan least squares (A^T A)^-1 A^T y
coeffs = np.linalg.lstsq(A, y, rcond=None)[0]
print(f"Koefisien hasil regresi:")
print(f"  Intercept : {coeffs[0]:.4f} (nilai asli: 1.0)")
print(f"  Slope     : {coeffs[1]:.4f} (nilai asli: 2.5)")
print()
print("Ini semua dikerjakan melalui operasi matriks!")
print("Di chapter-chapter berikutnya kita akan memahami semua ini dari dasar.")
````

````python
import matplotlib.pyplot as plt

# Visualisasi hasil regresi
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Scatter plot dengan garis regresi
axes[0].scatter(x, y, color="steelblue", edgecolors="k", s=30, alpha=0.7, label="Data")
y_pred = A @ coeffs
axes[0].plot(x, y_pred, color="tomato", linewidth=2.5, label=f"Regresi: y = {coeffs[1]:.2f}x + {coeffs[0]:.2f}")
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")
axes[0].set_title("Regresi Linear: Aplikasi Aljabar Linear")
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Visualisasi matriks desain
axes[1].imshow(A[:20], cmap="Blues", aspect="auto")
axes[1].set_title("Matriks Desain A (20 baris pertama)")
axes[1].set_xlabel("Kolom")
axes[1].set_ylabel("Baris (sampel)")
axes[1].set_xticks([0, 1])
axes[1].set_xticklabels(["Ones (bias)", "x"])
plt.colorbar(axes[1].images[0], ax=axes[1])

plt.tight_layout()
plt.show()

print("Aljabar linear memungkinkan kita memecahkan masalah regresi secara elegan.")
````

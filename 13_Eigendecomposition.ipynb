## Ringkasan Materi
Eigendecomposition adalah proses dekomposisi matriks menjadi vektor eigen (arah utama transformasi) dan nilai eigen (magnitudo perubahan). Ini adalah mesin matematis di balik analisis statistik seperti PCA, teknik reduksi dimensi, dan deteksi anomali.

### Topik yang Dibahas:
1. Interpretasi Geometris dan Statistika
2. Matriks Simetris dan Sifat Ortogonalnya
3. Quadratic Form dan Definiteness
4. Generalized Eigendecomposition


```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Dasar Eigendecomposition
A = np.array([[3, 1],
              [1, 3]])
evals, evecs = np.linalg.eig(A)

print("Nilai Eigen:", evals)
print("Vektor Eigen:\n", evecs)
```

***
## Matriks Simetris dan Definiteness

### Penjelasan Teori
Matriks simetris ($A = A^T$) memiliki keistimewaan karena semua vektor eigennya bersifat ortogonal dan nilai eigennya pasti bilangan real. Konsep **Definiteness** (Positive/Negative Definite) sangat penting dalam optimasi untuk menentukan apakah sebuah titik adalah minimum, maksimum, atau titik pelana (saddle point).



```python
# 2. Quadratic Form: f(x) = x^T * A * x
x = np.array([1, 1])
quad_form = x.T @ A @ x

print(f"Nilai Quadratic Form: {quad_form}")

# Definiteness (Positive Definite jika semua nilai eigen > 0)
is_pos_def = np.all(evals > 0)
print("Apakah matriks Positive Definite?", is_pos_def)
```

## Generalized Eigendecomposition

### Penjelasan Teori
Generalized eigendecomposition menyelesaikan persamaan $Av = \lambda Bv$. Ini digunakan ketika kita ingin memahami transformasi relatif antara dua matriks (misalnya dalam masalah diskriminan). Matriks $A^T A$ selalu *positive semi-definite*, yang menjamin stabilitas dalam banyak algoritma optimasi.

```python
B = np.array([[1, 0.5],
              [0.5, 1]])

# Menggunakan scipy untuk generalized eigenproblem
from scipy import linalg
evals_gen, evecs_gen = linalg.eig(A, B)

print("Generalized Eigenvalues:", evals_gen)
```

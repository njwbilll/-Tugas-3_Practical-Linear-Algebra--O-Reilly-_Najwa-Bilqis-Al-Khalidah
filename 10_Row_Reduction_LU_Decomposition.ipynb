## Ringkasan Materi
Bab ini mempelajari bagaimana komputer menyelesaikan sistem persamaan linear $Ax = b$. Kita mengeksplorasi teknik reduksi baris (row reduction) untuk mengubah matriks menjadi bentuk eselon baris, serta dekomposisi LU yang memecah matriks menjadi faktor segitiga bawah (Lower) dan segitiga atas (Upper).

### Topik yang Dibahas:
1. Konversi sistem persamaan menjadi matriks
2. Eliminasi Gaussian dan Gauss Jordan
3. Inversi matriks menggunakan reduksi baris
4. Dekomposisi LU dan matriks permutasi


```python
import numpy as np
import scipy.linalg as la

# 1. Sistem Persamaan Linear: Ax = b
A = np.array([[2, 1, 1],
              [1, 3, 2],
              [1, 0, 0]])
b = np.array([4, 5, 6])

# 2. Eliminasi Gaussian (Selesaikan Ax = b)
x = la.solve(A, b)
print("Solusi sistem persamaan x:", x)
```

***
## LU Decomposition

### Penjelasan Teori
Dekomposisi LU memecah matriks A menjadi A = P * L * U, di mana P adalah matriks permutasi (untuk menangani pertukaran baris), L adalah matriks segitiga bawah, dan U adalah matriks segitiga atas. Ini jauh lebih efisien daripada menghitung invers matriks secara langsung saat menyelesaikan banyak sistem dengan nilai b yang berbeda.



```python
# Melakukan LU Decomposition dengan pivoting (P)
P, L, U = la.lu(A)

print("Matriks Permutasi P:\n", P)
print("Matriks Segitiga Bawah L:\n", L)
print("Matriks Segitiga Atas U:\n", U)

# Verifikasi A = PLU
print("\nVerifikasi A = PLU:\n", np.round(P @ L @ U))
```

***
## Inversi melalui Gauss Jordan

### Penjelasan Teori
Metode Gauss Jordan melibatkan pembuatan matriks augmented [A | I], kemudian melakukan operasi baris elementer hingga sisi kiri menjadi matriks identitas I. Sisi kanan matriks augmented tersebut kemudian berubah menjadi invers dari A.




```python
# Membuat matriks augmented [A | I]
I = np.eye(len(A))
augmented = np.hstack((A, I))

print("Matriks Augmented [A | I]:\n", augmented)
# Komputasi reduksi baris secara manual dapat dilakukan dengan iterasi,
# namun di industri, kita menggunakan library scipy untuk stabilitas numerik.
```
